```python
def get_basic_bat(team, year,leauge):
    url2 = f"http://www.seamheads.com/NegroLgs/team.php?yearID={year}&teamID={team}&LGOrd={leauge}"

    df = pd.read_html(url2)[3].loc[:, ['Pos', 'Player', 'Age', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR',
       'RBI', 'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'DP', 'BA',
       'OBP', 'SLG', 'OPS', 'OPS+']].dropna(subset=["Pos"])
    df["team"] = team
    df["year"] = year
    df["leauge"] = leauge
    df["pit_bat"] = "bat"
    return df

def get_basic_pit(team, year, leauge):
    url2 = f"http://www.seamheads.com/NegroLgs/team.php?yearID={year}&teamID={team}&LGOrd={leauge}&tab=pit"

    df = pd.read_html(url2, attrs={'class' : 'main_table'})[1].loc[:, ['Pos', 'Player', 'Age', 'W', 'L', 'W%', 'G', 'GS', 'GF', 'CG', 'SHO',
       'SV', 'IP', 'R', 'ER', 'H', 'HR', 'BB', 'K', 'WP', 'BK', 'HBP', 'BF',
       'ERA', 'ERA+']].dropna(subset=["Pos"])
    df["team"] = team
    df["year"] = year
    df["leauge"] = leauge
    df["pit_bat"] = "pit"
    return df
```

```python
team_abbr = [
    "CAG", # Chicago American Giants
    "ABC", # Indianapolis ABCs
    "HIL", # hilldale Atheltic Club
    "HG", # Homestead Grays
    "PC", # Pitsburgh Crawfords
    "KCM", # KC Monarchs
    "BBB", # Birmingham Black Barons
    "CBE"  # Clevelend Buckeyes
]

years = [1916, 1922, 1924, 1931, 1936, 1938, 1944, 1945 ]
```

```python
df_bat = pd.concat([ get_basic_bat(team, year, lg) for team, year in zip(team_abbr, years) for lg in range(1,4) ])
df_pit = pd.concat([ get_basic_pit(team, year, lg) for team, year in zip(team_abbr, years) for lg in range(1,4) ])
```
